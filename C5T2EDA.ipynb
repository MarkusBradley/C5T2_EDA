```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

matrix = np.triu(credit.corr())
plt.figure(figsize=(20,15))
sns.heatmap(credit.corr(), annot = True, fmt='.2g', vmin=-1, vmax=1, center= 0, cmap= 'coolwarm', linewidths=0.5, linecolor='black', square=True, mask = matrix)
plt.show()
```

```python
# map features to their absolute correlation values
corr = credit.corr().abs()

# set equality (self correlation) as zero
# NOTE: Check if you have any variables that are collinear with the dependent variable first (this will not identify them)
corr[corr == 1] = 0

# of each feature, find the max correlation
# and sort the resulting array in ascending order
corr_cols = corr.max().sort_values(ascending=False)

# display the highly correlated features
display(corr_cols[corr_cols > 0.8])
```

# Discretizing Features

```python
import pandas as pd
credit = pd.read_csv('credit.csv')
def age_groups(series):
    if series < 30:
        return 20
    elif 30 <= series < 40:
        return 30
    elif 40 <= series < 50:
        return 40
    elif 50 <= series < 60:
        return 50
    elif 60 <= series:
        return 60

credit['AGE'] = credit['AGE'].apply(age_groups)

credit['AGE'].value_counts(sort = 0)
```

```python
def credit_groups(series):
    if series < 50000:
        return 50000
    elif 50000 <= series < 100000:
        return 100000
    elif 100000 <= series < 200000:
        return 200000
    elif 200000 <= series < 300000:
        return 200000
    elif 300000 <= series:
        return 300000
    
credit['LIMIT_BAL'] = credit['LIMIT_BAL'].apply(credit_groups)

credit['LIMIT_BAL'].value_counts(sort = 1)
```

```python
def edu_groups(series):
    if series < 1:
        return 0
    elif series == 1:
        return series
    elif series == 2:
        return series
    elif series == 3:
        return series
    elif series >= 4:
        return 0

    
credit['EDUCATION'] = credit['EDUCATION'].apply(edu_groups)

credit['EDUCATION'].value_counts(sort = 1)
```
